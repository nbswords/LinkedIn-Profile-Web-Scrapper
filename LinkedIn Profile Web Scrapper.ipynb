```python
import requests, time, random
from bs4 import BeautifulSoup
from selenium import webdriver
```

```python
# Download your corresponding driver from https://chromedriver.chromium.org/downloads
browser = webdriver.Chrome('driver/chromedriver.exe')
browser.get('https://www.linkedin.com/uas/login')

# Input your Linkedin account information
username = 'your linkedin username'
password = 'your linkedin password'

elementID = browser.find_element_by_id('username')
elementID.send_keys(username)

elementID = browser.find_element_by_id('password')
elementID.send_keys(password)

elementID.submit()
```

```python
# I take the profile of Databricks CEO as exmaple
link = 'https://www.linkedin.com/in/alighodsi/'
browser.get(link)
```

```python
SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = browser.execute_script("return document.body.scrollHeight")

for i in range(3):
    # Scroll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
src = browser.page_source
soup = BeautifulSoup(src, 'lxml')
print(soup)
```

## Personal information

```python
# Find all 'div' HTML tags in the top
name_div = soup.find('div', {'class': 'display-flex justify-space-between pt2'})
print(name_div)
```

### Find name

```python
where_is_name = name_div.find_all('h1')
print(where_is_name)
```

```python
# Because there is only one 'h1' tag so we just use .find()
name = name_div.find('h1').text.strip()
print(name)
```

### Find Locations

```python
where_is_loc = name_div.find_all('span')
print(where_is_loc)
```

```python
loc = soup.find('span',{'class':'text-body-small inline t-black--light break-words'}).text.strip()
print(loc)
```

### Find profile title

```python
where_is_title = name_div.find_all('div')
print(where_is_title)
```

```python
profile_title = soup.find('div',{'class':'text-body-medium break-words'}).text.strip()
print(profile_title)
```

### Find connection

```python
where_is_connection = name_div.find_all('span')
print(where_is_connection)
```

```python
connection = soup.find('span',{'class', 'dist-value'}).text.strip()
print(connection)
```

### Combined into personal linkedin information

```python
info = []
info.append(link)
info.append(name)
info.append(profile_title)
info.append(loc)
info.append(connection)
print(info)
```
